````python
exec(open('../src/create_csv.py').read())
from pybaseball import statcast
import time

for year in range(2024, 2025):
    try:
        # Fechas de inicio y fin del a√±o (para tener datos anuales)
        start_date = str(year) + '-01-01'
        end_date = str(year) + '-12-31'

        # Datos descargados por a√±o
        print(f"\n{'='*60}")
        print(f"Descargando datos de {start_date} a {end_date}...")
        print(f"{'='*60}")
        
        # Registrar tiempo de inicio
        inicio = time.time()
        
        # Descargar datos (esto puede tomar mucho tiempo)
        df = statcast(start_dt=start_date, end_dt=end_date)
        
        # Calcular tiempo transcurrido
        tiempo_descarga = time.time() - inicio
        
        # Guardar en CSV
        create_csv(df, start_date, end_date)

        # Informaci√≥n b√°sica de los datos descargados
        filas, columnas = df.shape
        print(f"Datos descargados: {filas:,} filas y {columnas} columnas")
        print(f"Tiempo de descarga: {tiempo_descarga:.2f} segundos")
        
        # Mostrar muestra de los datos
        print(f"\nPrimeras 3 filas:")
        print(df.head(3))
        
    except Exception as e:
        print(f"Error descargando datos para {year}: {e}")
        print("Los datos parciales deber√≠an estar guardados en cach√©")
        continue
````
