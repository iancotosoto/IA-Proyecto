# Datos de MLB con Python usando `pybaseball`

Esta secci√≥n resume las cuatro principales fuentes de datos accesibles a trav√©s de `pybaseball`, qu√© tipo de informaci√≥n trae cada una y c√≥mo obtenerlas f√°cilmente desde Python.

## üì¶ Instalaci√≥n

```bash
pip install pybaseball
1) Statcast (Baseball Savant) ‚Äî Pitch-by-pitch muy detallado
Qu√© datos trae:

Tipo de lanzamiento (pitch_type)

Velocidad (release_speed)

Spin (release_spin_rate)

Ubicaci√≥n en el plato (plate_x, plate_z)

Movimiento (pfx_x, pfx_z)

Resultado del lanzamiento (description)

Datos del contacto (launch_speed, launch_angle, estimated_woba_using_speedangle)

Uso t√≠pico: modelos para predecir ball/strike/swing, tipo de lanzamiento, calidad del contacto.

Ejemplo en Python:

python
Copy code
from pybaseball import statcast

df_statcast = statcast(start_dt="2024-04-01", end_dt="2024-04-02")
print(df_statcast.head())
2) FanGraphs ‚Äî M√©tricas avanzadas agregadas por jugador/equipo
Qu√© datos trae:

WAR, wOBA, wRC+, FIP, xFIP

Splits vs RHP / LHP, home/away

Proyecciones (ZiPS, Steamer)

Uso t√≠pico: an√°lisis de rendimiento, comparaci√≥n entre jugadores, modelos de predicci√≥n a nivel temporada.

Ejemplo:

python
Copy code
from pybaseball import fangraphs_leaders

df_fg = fangraphs_leaders(2024, stat="war", type="pit", qual=50)
print(df_fg.head())
3) Baseball-Reference ‚Äî Estad√≠sticas cl√°sicas y box scores
Qu√© datos trae:

Bateo (AVG, HR, RBI, OBP, SLG)

Pitcheo (ERA, WHIP, K%, BB%)

Box scores por juego

Resultados hist√≥ricos por temporada

Uso t√≠pico: an√°lisis de temporada, comparaci√≥n hist√≥rica, resumen de desempe√±o.

Ejemplo:

python
Copy code
from pybaseball import batting_stats, pitching_stats

df_bat_2024 = batting_stats(2024)
df_pit_2024 = pitching_stats(2024)
print(df_bat_2024.head())
print(df_pit_2024.head())
4) Retrosheet / Chadwick ‚Äî Play-by-play hist√≥rico + IDs
Qu√© datos trae:

Secuencia de jugadas (cada evento del partido)

Inning, outs, corredores en base, score en cada jugada

Mapeo de IDs de jugadores entre bases de datos

Uso t√≠pico: modelos de probabilidad de victoria (Win Probability), simulaciones, an√°lisis de estrategia.

Ejemplo (lookup de IDs):

python
Copy code
from pybaseball import playerid_lookup

ids = playerid_lookup("Trout", "Mike")
print(ids[["key_mlbam", "key_bbref", "key_retro"]])
üß≠ Resumen Comparativo
Fuente	Nivel	Tipo de datos	Ideal para
Statcast	Micro (lanzamiento)	F√≠sica + resultado	Modelos ML detallados (ball/strike/contacto)
FanGraphs	Macro avanzado	M√©tricas ajustadas	Evaluaci√≥n comparativa / proyecciones
Baseball-Reference	Macro cl√°sico	Estad√≠sticas resumidas	Tendencias hist√≥ricas y an√°lisis de temporada
Retrosheet	Micro secuencial	Play-by-play	Win Probability y simulaci√≥n de juego
